## IF STATEMENT WITH USER DEFINED FUNCTION

```python
# defining function
def price_label(row):
 if row['prices']<=5:
        return 'Low range product'
 elif(row['prices']>5 and row['prices']<=15):
        return 'Mid range product'
 elif row['prices']>15:
        return 'High range product'
 else: return 'Not enough data'
    
```

```python
# apply the function
df['price_range']=df.apply(price_label,axis=1)
```

```python
# to count
df['price_range'].value_counts(dropna=False)
```

```python
# check the maximum value in subset
df['prices'].max()
```

## STEP-4

```python
# count of orders per hour
df_orders_products_merged['order_hour_of_day'].value_counts(dropna=False)
```

### 24 hours of the day are divided into three parts:
### 'Most orders'= hour 10,11,14,15,13,12,16,9 i.e. from 9 am to 4 pm
### 'Average orders'= hour 17,8,18,19,20,7,21,22 i.e. from 7am to 8am and 5 pm to 10 pm
### 'Fewest orders'= hour 23,6,0,1,5,2,4,3 i.e. from 11pm to 5 am

```python
# create container
hours=[]
```

```python
for hour in df_orders_products_merged['order_hour_of_day']:
    if hour in [10,11,14,15,13,12,16,9]:
        hours.append('Most orders')
    elif hour in [17,8,18,19,20,7,21,22]:
        hours.append('Average orders')
    else:
        hours.append('Fewest orders')
```

```python
# create 'busiest_period_of_day' column in'df_orders_products_merged' data frame
df_orders_products_merged['busiest_period_of_day']=hours
```

```python
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna=False)
```
