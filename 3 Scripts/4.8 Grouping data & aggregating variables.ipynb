## STEP-4

```python
#AGGREGATING DATA WITH TRANSFORM
#1. first a new column 'max_order' is created to store the maximum order number of each user.
#2. then the data frame is groupby 'user_id' column
#3. finally transform function is applied to 'order_number' column
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# DERIVING COLUMNS WITH LOC
# if 'max_order' >40 then customers are 'loyal customer'
# if 'max_order' is <=40 and >10 then customers are 'regular customer'
# if 'max_order' <=10 then customers are 'new customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']>40,'loyalty_flag']='Loyal Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40) & (ords_prods_merge['max_order']>10),'loyalty_flag']='Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10,'loyalty_flag']='New Customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(20)
```

## STEP-5

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['min','max','mean','sum']})
```

### The Loyal customers purchase on average cheaper products than new and regular customers.

## STEP-6

```python
#first a new column'avg_price' is created to store average price of each user.
#then data frame is group by 'user_id'
# finally tranform function is applied to prices column
ords_prods_merge['avg_price']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
## creating a column 'spending_flag'to know 'low spender(<10)'and 'high spender(>=10)''
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price']<10,'spending_flag']='Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price']>=10,'spending_flag']='High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(10)
```

## STEP-7

```python
ords_prods_merge['median_frequency']=ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_frequency']>20,'customer']='Non frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_frequency']>10) & (ords_prods_merge['median_frequency']<=20),'customer']='Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_frequency']<=10,'customer']='Frequent customer'
```

```python
ords_prods_merge['customer'].value_counts(dropna=False)
```
